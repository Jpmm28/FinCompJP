Punto 2

$$0.5 x1 + 0.3 x2 + 0.25 x3 + 0.6 x4 ≥ 0.35 (valor)$$
$$0.3 x1 + 0.1 x2 + 0.4 x3 + 0.2 x4 ≥ 0.30 (tamaño)$$
$$0.2 x1 + 0.6 x2 + 0.35 x3 + 0.2 x4 ≥ 0.15 (liquidez)$$

$$x1 + x2 + x3 + x4 = $100$$
$$x1, x2, x3, x4 ≥ 0$$

Método Linprog

```python
c2 = np.array([0.1, 0.15, 0.16, 0.08])

A2d = np.array([[0.5, 0.3, 0.25, 0.6],
              [0.3, 0.1, 0.4, 0.2],
              [0.2, 0.6, 0.35, 0.2]
              ])

b2d = np.array([0.35, 0.30, 0.15])

A2i = ([[1,1,1,1]
        ])

b2i = np.array([1])

sol2 = linprog(-c2, A_ub=-A2d, b_ub=-b2d, A_eq=A2i, b_eq=b2i, method='simplex')
pd.DataFrame(sol2.x).round(4).T
```

Método CVXPY

```python
n=4
x2 = cp.Variable(n)
Obj = cp.Minimize(-c2.T@x2)
cons = [cp.sum(x2)==1,A2d@x2>=b2d,x2>=0]
prob =cp.Problem(Obj,cons)
Result = prob.solve()
x_sol2=pd.DataFrame(x2.value)
pd.DataFrame(x_sol2).round(4).T
```
